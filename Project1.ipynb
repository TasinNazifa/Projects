```python
import pandas as pd
```

```python
df = pd.read_csv("D:\\data\\Erin\\heart_disease_health_indicators.csv")
print(df)
```

```python
df.set_index("Age",inplace=True)
print(df)
```

```python
df.sort_index(inplace=True)
print(df)
```

```python
print(df.isnull().sum())
```

```python
print(df.describe())
print(df.info())
print(df.fillna(0,inplace=True))
```

```python
df.dropna(inplace=True) #to remove rows with missing values
print(df)
```

```python
df.dropna(axis=1,inplace=True) #To remove columns with missing values
print(df)
```

```python
df.drop_duplicates(inplace=True)
print(df)
```

```python
df['Sex']=df['Sex'].map({0:'Female',1:'Male'})
print(df)
```

```python
Q1 = df['MentHlth'].quantile(0.25)
Q3 = df['MentHlth'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['MentHlth'] >= (Q1 - 1.5 * IQR)) & (df['MentHlth'] <= (Q3 + 1.5 * IQR))]
print(df)
```

```python
Q1 = df['PhysHlth'].quantile(0.25)
Q3 = df['PhysHlth'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['PhysHlth'] >= (Q1 - 1.5 * IQR)) & (df['PhysHlth'] <= (Q3 + 1.5 * IQR))]
print(df)
```

```python
Q1 = df['GenHlth'].quantile(0.25)
Q3 = df['GenHlth'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['GenHlth'] >= (Q1 - 1.5 * IQR)) & (df['GenHlth'] <= (Q3 + 1.5 * IQR))]
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
grouped_data = df.groupby('Age')['MentHlth'].mean().reset_index()
grouped_data2 = df.groupby('Age')['PhysHlth'].mean().reset_index()
grouped_data3 = df.groupby('Age')['GenHlth'].mean().reset_index()
grouped_data7 = df.groupby('Age')['HvyAlcoholConsump'].mean().reset_index()
plt.figure(figsize=(12, 4))
plt.plot(grouped_data['Age'], grouped_data['MentHlth'], marker='o', linestyle='-', color='blue',label="MentalHealth")
plt.plot(grouped_data2['Age'],grouped_data2['PhysHlth'],marker='o',color='green',label="PhysicalHealth")
plt.plot(grouped_data3['Age'],grouped_data3['GenHlth'],marker='o',color='purple',label="GeneralHealth")
plt.plot(grouped_data7['Age'],grouped_data7['HvyAlcoholConsump'],marker='o',color='red',label="HeavyAlcoholConsumption")
plt.xlabel('Age')
plt.ylabel('Average Poor Mental,Physical,General Health Days')
plt.title('Age vs Mental,Physical,General Health')
plt.grid(True)
plt.legend()
plt.show()
```
